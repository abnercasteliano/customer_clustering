# Customer Segmentation - Machine Learning

```python
# principais módulos
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# importando o DataSet salvo durante Data Wrangling e EDA

df = pd.read_csv('customer_data2.csv')
df.head()
```

## Feature Scaling/Normalization

**Padronização dos dados em uma escala comum.**

```python
# importando standard scaler

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# Definindo dados de Entrada - INPUT

X = df.values

scaled_X = scaler.fit_transform(X)
```

## Aplicando a redução de dimensionalidade PCA

**Como o dataset é multidimensional, o algoritmo irá performar melhor utilizando apenas os componentes principais.**

```python
# importando o PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2).fit_transform(scaled_X)

# criando um novo daframe com os componentes principais
pca_df = pd.DataFrame(data=pca, columns=['PCA1', 'PCA2'])
```

```python
pca_df.head()
```

## Avaliando o melhor modelo de clusterização

### K-means

```python
# importando o algoritmo Kmeans

from sklearn.cluster import KMeans
```

```python
# Definindo o melhor valor de K para o conjunto de dados- Elbow Method

# Inertia - distância dos pontos de dados até o centro do seu cluster
inertia = []
# Limites de K
k_range = range(1, 15)

for i in k_range:
    kmeans = KMeans(n_clusters=i)
    kmeans.fit_predict(pd.DataFrame(scaled_X))
    inertia.append(kmeans.inertia_)
plt.figure(figsize=(8,6))
plt.plot(k_range, inertia, 'bx-')
plt.xlabel('K Values')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

**Neste caso o melhor número de K que gera a melhor inertia é o 4.**

```python
# criando o modelo K means com 4 clusters 

kmeans_model = KMeans(4)
# aplicando a clusterização
kmeans_model.fit_predict(pca)

# concatenando os valores previstos no agrupamento com os componentes principais
# para facilitar a criação dos gráficos
pca_df_kmeans = pd.concat([pca_df, pd.DataFrame({'CLUSTER':kmeans_model.labels_})], axis=1)
```

```python
# Plot de cada cluster
plt.figure(figsize=(10,8))
ax = sns.scatterplot(x='PCA1', y='PCA2',data = pca_df_kmeans, hue= 'CLUSTER', palette=['red', 'green', 'blue', 'purple'])
# plot das centróides
ax = sns.scatterplot(x=kmeans_model.cluster_centers_[:, 0], y=kmeans_model.cluster_centers_[:, 1],
                     hue=range(4), palette=['red', 'green', 'blue', 'purple'], s=169, ec='black', legend=False, ax=ax)
plt.title('K-Means')
plt.show()
```

### Avaliando o modelo - Silhouette Score

**O Silhouette Score, descreve o quão bem os dados estão agrupados corretamente.**

**Valores próximos de 1 são melhores, sendo 1 o melhor resultado possível.**

**Valores negativos indicam que a amostra foi agrupada de forma errada, sendo -1 o pior valor possível.**

**Valores perto de 0 ou 0 indicam Overlapping. Um mesmo dado pertencendo a diferentes clusters, o que dificulta a tomada de decisão.**

```python
# computando silhouette score
from sklearn.metrics import silhouette_score

labels = kmeans_model.labels_

silhouette_score(pca, labels, metric='euclidean')
```

### DBSCAN

**O algoritmo DBSCAN, não precisa de um número de clusters pré definido. Porém devemos definir outros parâmetros importantes.**

```python
from sklearn.cluster import DBSCAN
# criando o modelo e aplicando clusterização
dbscan = DBSCAN(eps=0.5, min_samples = 5, metric = 'euclidean')
dbscan.fit_predict(pca)

# concatenando os valores previstos no agrupamento com os componentes principais
# para facilitar a criação dos gráficos
pca_dbscan = pd.concat([pca_df, pd.DataFrame({'CLUSTER':dbscan.labels_})], axis=1)
```

```python
# Plot de cada cluster
plt.figure(figsize=(10,8))
ax = sns.scatterplot(x='PCA1', y='PCA2',data = pca_dbscan, hue= 'CLUSTER', palette='bright')
# plot das centróides
plt.title('DBSCAN')
plt.show()
```

**O DBSCAN identificou 4 clusters um grupo grande de ruídos.**

```python
# criando um data Set final para armazenar os dados ruidosos e tratá-los separadamente

noisy_data = pd.concat([df, pd.DataFrame({'Cluster':dbscan.labels_})], axis=1)
```

```python
# coletando os dados ruidosos

noisy_data[noisy_data['Cluster'] == -1]
```

```python
# computando silhouette score
from sklearn.metrics import silhouette_score

labels = dbscan.labels_

silhouette_score(pca, labels, metric='euclidean')
```

**O DBSCAN apresentou uma performance melhor, mas vamos utilizar o k mean, pois o foco deste problemas é definir o melhor número de clusters e não identificar ruídos.**

## Agglomerative Clustering

**Método de clusterização hierárquico.**

```python
# aglomerative clustering

from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
```

```python
# realizando os agrupamentos

hc.fit_predict(pca)

# concatenando os valores previstos no agrupamento com os componentes principais
# para facilitar a criação dos gráficos
pca_hc = pd.concat([pca_df, pd.DataFrame({'CLUSTER':hc.labels_})], axis=1)
```

```python
# Plot de cada cluster
plt.figure(figsize=(10,8))
ax = sns.scatterplot(x='PCA1', y='PCA2',data = pca_hc, hue= 'CLUSTER', palette="bright")
# plot das centróides
plt.title('Agglomerative Clustering')
plt.show()
```

```python
# computando silhouette score
from sklearn.metrics import silhouette_score

labels = hc.labels_

silhouette_score(pca, labels, metric='euclidean')
```

**O Agglomerative clustering apresentou performance inferior em relação ao K means.**

## Mean Shift

**Técnica Particional de clusterização. Utiliza a função de densidade de Kernel para definir os clusters.**

```python
from sklearn.cluster import MeanShift

mean_shift = MeanShift()
```

```python
# realizando os agrupamentos

mean_shift.fit_predict(pca)

# concatenando os valores previstos no agrupamento com os componentes principais
# para facilitar a criação dos gráficos
pca_mean_sft = pd.concat([pca_df, pd.DataFrame({'CLUSTER':mean_shift.labels_})], axis=1)
```

```python
# Plot de cada cluster
plt.figure(figsize=(10,8))
ax = sns.scatterplot(x='PCA1', y='PCA2',data = pca_mean_sft, hue= 'CLUSTER', palette="bright")
# plot das centróides
plt.title('Mean Shift')
plt.show()
```

```python
# computando silhouette score
from sklearn.metrics import silhouette_score

labels = mean_shift.labels_

silhouette_score(pca, labels, metric='euclidean')
```

**O Mean Shift teve boa performance, mas gerou muitos clusters repetidos.**

## Spectral Clustering

**Técnica particional avançada, o spectral utiliza grafos, autovalor e autovetor para definir os grupos.**

```python
# Spectral Clustering

from sklearn.cluster import SpectralClustering

spectral = SpectralClustering(n_clusters=4, assign_labels='discretize', random_state=0)
```

```python
# realizando os agrupamentos

spectral.fit_predict(pca)

# concatenando os valores previstos no agrupamento com os componentes principais
# para facilitar a criação dos gráficos
pca_spectral = pd.concat([pca_df, pd.DataFrame({'CLUSTER':spectral.labels_})], axis=1)
```

```python
# Plot de cada cluster
plt.figure(figsize=(10,8))
ax = sns.scatterplot(x='PCA1', y='PCA2',data = pca_spectral, hue= 'CLUSTER', palette="bright")
# plot das centróides
plt.title('Spectral Clustering')
plt.show()
```

```python
# computando silhouette score
from sklearn.metrics import silhouette_score

labels = spectral.labels_

silhouette_score(pca, labels, metric='euclidean')
```

**O Spectral Cluster retornou uma performance melhor em relação aos outros.**

**Insight** - Apesar da performance inferior em relação a outros modelos, o k means apresentou uma segmentação de clientes mais distribuída e diversificada.

**O modelo final escolhido para o problema será o K means.**

# Criando um modelo de classificação a partir dos dados clusterizados

**Agora que o algoritmo identificou os grupos, podemos criar um modelo supervisionado para classificar novos clientes em diferentes grupos.**

**Este processo evita ter que repetir todo o processo de definição dos grupos em novos dados.**

```python
final_df.shape
```

```python
# dividindo os dados em INPUT e OUTPUT

array = final_df.values

X = array[:, 0:17]
y = array[:,17]
```

## Feature Selection

```python
# Feature Selection utilizando Extra Trees Classifier

from sklearn.ensemble import ExtraTreesClassifier

# instanciando o modelo
extra_trees = ExtraTreesClassifier()
extra_trees.fit(X, y)

# printando os resultados

print(extra_trees.feature_importances_)
```

```python
# printando os resultados maiores

variables = df.columns[0:17]

print(variables[0], variables[2], variables[5], variables[6], variables[7], variables[8],
      variables[9], variables[10], variables[11], variables[12])
```

## Coletando o array final com as features importances

```python
# Final INPUT
X = X[:,[0, 2, 5, 6, 7, 8, 9, 10, 11, 12]]

# Final OUTPUT
y = y
```

## Dividindo os dados de treino e teste

```python
# Dividindo os dados de treino e teste 

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state=7)
```

## Ridge Classifier model

```python
from sklearn.linear_model import RidgeClassifier
ridge_class = RidgeClassifier(alpha=1)
from sklearn.metrics import classification_report
```

```python
# Treinando e avaliando o modelo

ridge_class.fit(X_train, y_train)

y_pred = ridge_class.predict(X_test)
```

```python
print('\n','Classification Report','\n')
print(classification_report(y_test, y_pred))
```

**Considerando F1_score macro: 78% e acurácia : 85%**

## Decision Tree Classifier

```python
### Ajuste de hiperparâmetros para o modelo

from sklearn.model_selection import RandomizedSearchCV

# profundidade máxima da árvore limite de parada de nós
max_depth = [int(x) for x in np.linspace(3, 100, 10)]

# máximo de variáveis por nó
max_features = ['auto', 'sqrt', 'log2']

# número mínimo de amostras por nó split
min_samples_split = [2, 3, 4, 5, 8, 9]

# número mínimo de amostras por folha
min_samples_leaf = [1, 2, 3, 4, 6, 8]

# critério de divisão
criterion = ['entropy', 'gini']

random_grid = {'max_depth' : max_depth, 
               'max_features' : max_features, 
               'criterion' : criterion, 
               'min_samples_leaf' : min_samples_leaf, 
               'min_samples_split' : min_samples_split}
```

**O ajuste é muito importante na árvore de decisão pois cada conjunto de dados possui suas particularidades.**

```python
# iniciando e treinando o modelo com Randomized Grid Search

from sklearn.tree import DecisionTreeClassifier
dt_classifier = DecisionTreeClassifier()

dt_randGrid = RandomizedSearchCV(estimator=dt_classifier, param_distributions=random_grid, 
                                 n_iter=100, cv=3, random_state = 7)

dt_randGrid.fit(X_train, y_train)
```

```python
# verificando os melhores parâmetros
dt_randGrid.best_params_
```

```python
# Criando e testando o modelo novamente

dt_classifier = DecisionTreeClassifier(min_samples_split = 9,
                                       min_samples_leaf = 4,
                                       max_features = 'log2',
                                       max_depth = 46,
                                       criterion = 'entropy')

dt_classifier.fit(X_train, y_train)

y_pred = dt_classifier.predict(X_test)
```

```python
# avalidando o resultado do modelo

print('\n','Classification Report','\n')
print(classification_report(y_test, y_pred))
```

**O Decision Tree Classifier aprensentou um F1_Score macro de 87% e uma acurácia de 89%.**

## Random Forest Classifier

```python
# Ajustando os parâmetros

from sklearn.model_selection import RandomizedSearchCV

# numero de árvores para o random forest
n_estimators= [int(x) for x in np.linspace(start=100, stop = 1000, num = 10)]

# profundidade máxima de cada árvore
max_depth = [int(x) for x in np.linspace(3, 100, 10)]

# máximo de variáveis por nó
max_features = ['auto', 'sqrt', 'log2']

# número mínimo de amostras por nó split
min_samples_split = [2, 3, 4, 5, 8, 9]

# número mínimo de amostras por folha
min_samples_leaf = [1, 2, 3, 4, 6, 8]

# critério de divisão
criterion = ['entropy', 'gini']

random_grid = {'n_estimators': n_estimators,
               'max_depth' : max_depth, 
               'max_features' : max_features, 
               'criterion' : criterion, 
               'min_samples_leaf' : min_samples_leaf, 
               'min_samples_split' : min_samples_split}
```

```python
# criando o modelo com Search CV

from sklearn.ensemble import RandomForestClassifier

rf_classifier = RandomForestClassifier()

rf_randGrid = RandomizedSearchCV(estimator=rf_classifier, param_distributions=random_grid, 
                                 n_iter=100, cv=3, random_state = 7, n_jobs=-1)

rf_randGrid.fit(X_train, y_train)
```

```python
# verificando os melhores parâmetros
rf_randGrid.best_params_
```

```python
# Criando e testando o modelo novamente

rf_classifier = RandomForestClassifier(n_estimators = 500,
                                       min_samples_split = 2,
                                       min_samples_leaf = 1,
                                       max_features = 'auto',
                                       max_depth = 100,
                                       criterion = 'gini')

rf_classifier.fit(X_train, y_train)

y_pred = rf_classifier.predict(X_test)
```

```python
# avaliando o resultado do modelo

print('\n','Classification Report','\n')
print(classification_report(y_test, y_pred))
```

**O Random Forest Classifier aprensentou a melhor performance, com um F1_Score macro de 92% e uma acurácia de 93%.**
